```python
# Exemplo de uso de JSON no livro Python Fluente

from urllib.request import urlopen
import warnings
import os
import json

url_fluent = 'https://www.oreilly.com/pub/sc/osconfeed'
JSON = 'data/osconfeed.json'
```

```python
def load():
    if not os.path.exists(JSON):
        msg = 'Downloading {} to {}'.format(url_fluent, JSON)
        warnings.warn(msg)
        with urlopen(url_fluent) as remote, open(JSON, 'wb') as local:
            local.write(remote.read())
    
    with open(JSON) as fp:
        return json.load(fp)
```

```python
feed = load()
```
